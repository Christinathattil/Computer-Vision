```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import imageio

```

```python
# Define image file paths
input_image_path = "/content/cv_image .jpeg"
output_image_path = "/content/Garnter_HypeCycle_23a.png"

```

```python
# Read image using OpenCV (cv2)
image_cv2 = cv2.imread(input_image_path)


# Display image using OpenCV (cv2)
#cv2.imshow("Image (OpenCV)", image_cv2)
from google.colab.patches import cv2_imshow
cv2.waitKey(0)
cv2.destroyAllWindows()

# Write image using OpenCV (cv2)
cv2.imwrite(output_image_path + "_cv2.jpg", image_cv2)
```

```python
# Read image using PIL (Python Imaging Library)
image_pil = Image.open(input_image_path)

# Display image using PIL (Python Imaging Library)
image_pil.show()

# Write image using PIL (Python Imaging Library)
image_pil.save(output_image_path + "_pillow.jpg")
```

```python
# Read image using Matplotlib
image_matplotlib = plt.imread(input_image_path)

# Display image using Matplotlib
plt.imshow(image_matplotlib)
plt.axis('off')
plt.show()

# Write image using Matplotlib
plt.imsave(output_image_path + "_matplotlib.jpg", image_matplotlib)
```

```python
# Read image using ImageIO
image_imageio = imageio.imread(input_image_path)

# Display image using ImageIO
from google.colab.patches import cv2_imshow
plt.imshow(image_imageio)

# Write image using ImageIO
imageio.imwrite(output_image_path + "_imageio.jpg", image_imageio)
```
